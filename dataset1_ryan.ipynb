```python
# Distribution of resale prices
sns.histplot(df['resale_price'], kde=True)
plt.title('Distribution of Resale Prices')
plt.show()

# Distribution of floor area
sns.histplot(df['floor_area_sqm'], kde=True)
plt.title('Distribution of Floor Area (sqm)')
plt.show()

# Distribution of lease commence date
sns.histplot(df['lease_commence_date'], kde=True)
plt.title('Distribution of Lease Commence Date')
plt.show()
```

```python
# Scatter plot between floor area and resale price
plt.scatter(df['floor_area_sqm'], df['resale_price'])
plt.title('Floor Area vs. Resale Price')
plt.xlabel('Floor Area (sqm)')
plt.ylabel('Resale Price')
plt.show()

# Relationship between storey range and resale price
sns.boxplot(x='storey_range', y='resale_price', data=df)
plt.title('Storey Range vs. Resale Price')
plt.xticks(rotation=90)
plt.show()

# Relationship between flat type and resale price
sns.boxplot(x='flat_type', y='resale_price', data=df)
plt.title('Flat Type vs. Resale Price')
plt.show()

```

```python
# Scatter plot showing the resale prices on a geographical map
plt.figure(figsize=(10,8))
sns.scatterplot(x='longitude', y='latitude', hue='resale_price', palette='viridis', data=df)
plt.title('Geographical Distribution of Resale Prices')
plt.show()
```

```python
# Distribution of distance to CBD
sns.histplot(df['cbd_dist'], kde=True)
plt.title('Distribution of Distance to CBD')
plt.show()

# Relationship between distance to CBD and resale price
sns.scatterplot(x='cbd_dist', y='resale_price', data=df)
plt.title('Distance to CBD vs. Resale Price')
plt.show()

# Relationship between distance to nearest MRT and resale price
sns.scatterplot(x='min_dist_mrt', y='resale_price', data=df)
plt.title('Distance to MRT vs. Resale Price')
plt.show()
```
